```python

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import LabelEncoder

# Assuming df2 is your DataFrame
df = df2.copy()

# Drop non-numeric columns (like 'Record' which contains strings)
df = df.drop(columns=["Record"], errors='ignore')  # Remove 'Record' safely if it exists

# Ensure all categorical columns are converted to numbers
categorical_cols = df.select_dtypes(include=['object']).columns  # Find categorical columns
for col in categorical_cols:
    df[col] = LabelEncoder().fit_transform(df[col])  # Convert to numerical values

# Splitting into features (X) and target (Y)
X = df.iloc[:, :-3]  # Selecting features till 'Samp. en.'
Y = df["Premature"]  # Selecting target

# Verify no mismatch
print(X.shape, Y.shape)  # They should match

# Train-test split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

# Train Logistic Regression model
clf = LogisticRegression()
clf.fit(X_train, Y_train)

# Predict and check accuracy
from sklearn.metrics import accuracy_score
Y_pred = clf.predict(X_test)
print("Accuracy:", accuracy_score(Y_test, Y_pred))
```
