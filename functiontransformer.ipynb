```python
import numpy as np 
import pandas as pd 
from sklearn.model_selection import train_test_split 
from sklearn.preprocessing import FunctionTransformer 
import scipy.stats as stats 
from sklearn.tree import DecisionTreeRegressor
from sklearn.linear_model import LinearRegression
from sklearn.compose import ColumnTransformer
import seaborn as sns 
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error
df=pd.read_csv("concrete_data.csv")
X=df.iloc[:,0:8]
Y=df.iloc[:,-1]
[X_train,X_test,Y_train,Y_test]=train_test_split(X,Y)
for col in X_train.columns:
    plt.figure(figsize=(14,4))
    plt.subplot(121)
    sns.distplot(X_train[col])
    plt.title("col")
    plt.subplot(122)
    stats.probplot(X_train[col],dist="norm",plot=plt)
    plt.title("col")
    plt.show()
```

```python
clf=LinearRegression()
tree=DecisionTreeRegressor()
clf.fit(X_train,Y_train)
Y_pred=clf.predict(X_test)
tree.fit(X_train,Y_train)
Y_pred2=tree.predict(X_test)
print("mean squared error of clf",mean_squared_error(Y_pred,Y_test))
print("mean squared error of tree",mean_squared_error(Y_pred2,Y_test))
```

```python
transform=FunctionTransformer(func=np.log1p)
X_train_transformed=transform.fit_transform(X_train)
X_test_transformed=transform.transform(X_test)
clf.fit(X_train_transformed,Y_train)
Y_pred3=clf.predict(X_test_transformed)
tree.fit(X_train_transformed,Y_train)
Y_pred4=tree.predict(X_test_transformed)
print("mean squared error after function transformer clf",mean_squared_error(Y_pred3,Y_test))
print("mean squared error after function transformer tree",mean_squared_error(Y_pred4,Y_test))
```
