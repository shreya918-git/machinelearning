```python
import requests
import pandas as pd
import time

df = pd.DataFrame()  # Initialize an empty DataFrame

for i in range(1, 429):
    try:
        # Send the request
        response = requests.get(
            "https://api.themoviedb.org/3/discover/movie?api_key=b780cd87f8ca7aba9258415af9c71730"
            "&include_adult=false&include_video=false&language=en-US&page={}&sort_by=vote_average.desc"
            "&without_genres=99,10755&vote_count.gte=200".format(i)
        )
        
        # Check if the response is valid
        if response.status_code == 200 and "results" in response.json():
            # Convert results to DataFrame
            temp_df = pd.DataFrame(response.json()["results"])[[
                "id", "original_language", "original_title", "overview", 
                "popularity", "release_date", "title", "vote_average", "vote_count"
            ]]
            
            # Concatenate to main DataFrame
            df = pd.concat([df, temp_df], ignore_index=True)
        else:
            print(f"Error: Status code {response.status_code} on page {i}")
    except Exception as e:
        print(f"Error on page {i}: {e}")
    
    # Pause to avoid rate limits
    time.sleep(0.5)

# Display the DataFrame summary
print(df.head())
print(f"Total rows: {df.shape[0]}")

# Optionally save the DataFrame to a CSV file
df.to_csv("tmdb_movies.csv", index=False)
```
