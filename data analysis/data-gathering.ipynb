```python
response=requests.get("https://www.ambitionbox.com/list-of-companies?campaign=desktop_nav",headers=headers).text
```

```python
soup=BeautifulSoup(response,"lxml")
companies=[]
ratings=[]
reviews=[]
ctype=[]
jobs=[]
salaries=[]
for i in soup.find_all("div",class_="companyCardWrapper"):
    companies.append(i.find("h2",class_="companyCardWrapper__companyName").text.strip())
    ratings.append(i.find("div",class_="rating_text rating_text--md").text.strip())
    reviews.append(i.find_all("span",class_="companyCardWrapper__ActionCount")[0].text.strip())
    ctype.append(i.find("span",class_="companyCardWrapper__interLinking").text.strip())
    jobs.append(i.find_all("span",class_="companyCardWrapper__ActionCount")[1].text.strip())
    salaries.append(i.find_all("span",class_="companyCardWrapper__ActionCount")[2].text.strip())

df=pd.DataFrame({"companies":companies,"ratings":ratings,"reviews":reviews,"company_type":ctype,"jobs":jobs,"salaries":salaries})
df.to_csv("companies.csv",index=False)
```
