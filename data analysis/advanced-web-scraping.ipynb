```python
import numpy as np 
import pandas as pd 
from bs4 import BeautifulSoup
```

```python
with open("smartprix.html","r") as f:
    html=f.read()
```

```python
soup=BeautifulSoup(html,"lxml")
names=[]
prices=[]
ratings=[]
sim=[]
processor=[]
ram=[]
battery=[]
display=[]
camera=[]
card=[]
os=[]
containers=soup.find_all("div",class_="sm-product has-tag has-features has-actions")
for i in containers:
    names.append(i.find("h2").text.strip())
    prices.append(i.find("span",class_="price").text.strip())
    try:
        ratings.append(i.find("div",class_="score rank-2-bg").find("b").text.strip())
    except:
        ratings.append(np.nan)
    try:
        sim.append(i.find("ul",class_="sm-feat specs").find_all("li")[0].text.strip())
    except:
        sim.append(np.nan)
    try:
        processor.append(i.find("ul",class_="sm-feat specs").find_all("li")[1].text.strip())
    except:
        processor.append(np.nan)
    try:
        ram.append(i.find("ul",class_="sm-feat specs").find_all("li")[2].text.strip())
    except:
        ram.append(np.nan)
    try:
        battery.append(i.find("ul",class_="sm-feat specs").find_all("li")[3].text.strip())
    except:
        battery.append(np.nan)
    try:
        display.append(i.find("ul",class_="sm-feat specs").find_all("li")[4].text.strip())
    except:
        display.append(np.nan)
    try:
        camera.append(i.find("ul",class_="sm-feat specs").find_all("li")[5].text.strip())
    except:
        camera.append(np.nan)
    try:
        card.append(i.find("ul",class_="sm-feat specs").find_all("li")[6].text.strip())
    except:
        card.append(np.nan)
    try:
        os.append(i.find("ul",class_="sm-feat specs").find_all("li")[7].text.strip())
    except:
        os.append(np.nan)
df=pd.DataFrame({"names":names,"prices":prices,"ratings":ratings,"sim":sim,"processor":processor,"ram":ram,"battery":battery,"display":display,"camera":camera,"card":card,"os":os})
df=df.drop_duplicates(keep="first")
df.to_excel("smartphones.xlsx",index=False)
```
