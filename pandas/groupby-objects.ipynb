```python
import numpy as np 
import pandas as pd 
movies=pd.read_csv("imdb-top-1000.csv")
genre=movies.groupby("Genre")
genre.min()
genre.max()
genre.sum()
genre.describe()
# genre.mean()
genre.sum().sort_values("No_of_Votes",ascending=False)
genre["Gross"].sum().sort_values(ascending=False).head(3)
genre["IMDB_Rating"].mean().sort_values(ascending=False).head(3) 
movies.groupby("Director")["No_of_Votes"].sum().sort_values(ascending=False).head(3)
movies.groupby("Star1")["Series_Title"].count().sort_values(ascending=False)
len(genre)
genre.size()
genre.first()
genre.last()
genre.nth(7)
genre.get_group("Horror")
genre.groups
genre.sample(2,replace=True)
genre.nunique() 
genre.agg({"Series_Title":["min","max"],"IMDB_Rating":"mean"})
genre.agg(["sum"])
```

```python
df=pd.DataFrame(columns=movies.columns)
for i,j in genre:
    df=pd.concat((df,j[j["IMDB_Rating"]==j["IMDB_Rating"].max()]))
df
```

```python
def starts_with_a(group):
    return group["Series_Title"].str.startswith("A").sum() 
genre.apply(starts_with_a)
```

```python
def rank_movies(group):
    group["rank"]=group["IMDB_Rating"].rank(ascending=False)
    return group 
genre.apply(rank_movies)
```

```python
def normal(group):
    group["normal"]=(group["IMDB_Rating"]-group["IMDB_Rating"].min())/(group["IMDB_Rating"].max()-group["IMDB_Rating"].min())
    return group 
genre.apply(normal)
```
