```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
iris=pd.read_csv("iris.csv")
iris
```

```python
iris["Species"]=iris["Species"].replace({"Iris-setosa":1,"Iris-virginica":2,"Iris-versicolor":3})
plt.scatter(iris["SepalLengthCm"],iris["PetalLengthCm"],c=iris["Species"],cmap="viridis",alpha=0.7)
plt.colorbar()
```

```python
batter=pd.read_csv("batter.csv")
plt.figure(figsize=(15,6))
plt.scatter(batter["avg"],batter["strike_rate"])
for i in range(0,25):
    plt.text(batter["avg"].values[i],batter["strike_rate"].values[i],batter["batter"].values[i],fontdict={"color":"red"})
plt.axhline(130,color="red")
plt.axhline(140)
plt.axvline(40)
```

```python
fig,ax=plt.subplots(nrows=2,ncols=1,sharex=True)
ax[0].scatter(batter["avg"],batter["strike_rate"],color="red")
ax[0].set_title("avg vs strike rate")
ax[0].set_ylabel("strike rate")
ax[1].scatter(batter["avg"],batter["runs"])
ax[1].set_title("avg vs runs")
ax[1].set_xlabel("avg")
ax[1].set_ylabel("runs")
```

```python
fig,ax=plt.subplots(nrows=2,ncols=2,figsize=(15,10))
ax[0,0].scatter(batter["avg"],batter["strike_rate"])
ax[0,1].scatter(batter["avg"],batter["runs"])
ax[1,0].hist(batter["avg"])
ax[1,1].hist(batter["runs"])
```

```python
fig=plt.figure()
ax1=fig.add_subplot(2,2,1)
ax1.scatter(batter["avg"],batter["strike_rate"])
ax2=fig.add_subplot(2,2,2)
ax2.scatter(batter["avg"],batter["runs"])
```

```python
fig=plt.figure()
ax=plt.subplot(projection="3d")
ax.scatter3D(batter["avg"],batter["strike_rate"],batter["runs"])
ax.set_title("IPL Batsman analysis")
```

```python
x=[10,20,30,40]
y=[0,20,30,40]
z=[0,20,30,50]
ax1=plt.subplot(projection="3d")
ax1.scatter(x,y,z,marker="+",s=[100,100,100,100],alpha=0.7)
ax1.plot3D(x,y,z,color="red")
```

```python
x=np.linspace(10,-10,100)
y=np.linspace(10,-10,100)
xx,yy=np.meshgrid(x,y)
zz=np.sin(xx)+np.cos(yy)
ax2=plt.subplot(projection="3d")
p=ax2.plot_surface(xx,yy,zz,cmap="viridis")
fig.colorbar(p)
```

```python
ax3=plt.subplot()
ax3.contour(xx,yy,zz)
ax3.contourf(xx,yy,zz)
```

```python
ipl=pd.read_csv("IPL_Ball_by_Ball_2008_2022 (1).csv")
ipl=ipl[(ipl["ballnumber"].isin([1,2,3,4,5,6])) & (ipl["batsman_run"]==6)]
ipl2=ipl.pivot_table(index="overs",columns="ballnumber",values="batsman_run",aggfunc="count")
plt.imshow(ipl2)
plt.yticks(sorted(ipl["overs"].unique()),list(range(1,21)))
plt.xticks(np.arange(0,6),list(range(1,7)))
plt.colorbar()
```

```python
tips=sns.load_dataset("tips")
tips["size"]=tips["size"]*100
tips.plot(kind="scatter",x="total_bill",y="tip",marker="+",s=tips["size"],c=tips["sex"],cmap="viridis")
```
