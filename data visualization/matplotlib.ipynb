```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
price=[48000,49000,50000,40000,35000,30000]
year=[2015,2016,2017,2018,2019,2020]
plt.plot(year,price)
```

```python
batsman=pd.read_csv("sharma-kohli.csv")
plt.plot(batsman["index"],batsman["V Kohli"],color="black",linestyle="solid",linewidth=2,marker="D",markersize=5,label="V Kohli")
plt.plot(batsman["index"],batsman["RG Sharma"],color="yellow",linestyle="dashed",linewidth=2,marker="+",markersize=5,label="RG Sharma")
plt.title("RG Sharma vs V Kohli")
plt.xlabel("Year")
plt.ylabel("Runs")
plt.legend(loc="best")
plt.grid()
plt.show()
```
