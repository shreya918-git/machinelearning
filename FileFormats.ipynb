```python
import pandas as pd 
# Create an example dataset with Year, Month, and Sales columns
data = {
    'Year': [2023, 2023, 2023, 2024, 2024, 2024, 2025, 2025],
    'Month': [1, 2, 3, 1, 2, 3, 1, 2],
    'Sales': [5000, 4500, 6000, 7000, 7500, 8000, 9000, 9500]
}

# Create a DataFrame from the data dictionary
df = pd.DataFrame(data)

# Specify the CSV file path
csv_file_path = 'monthly_sales.csv'

# Save the DataFrame to a CSV file
df.to_csv(csv_file_path, index=False)

# Read the CSV without calling info() directly (info() does not return the DataFrame)
df2 = pd.read_csv(csv_file_path)

# Combine 'Year' and 'Month' into a single string and convert it to datetime
df2['Date'] = pd.to_datetime(df2["Year"].astype(str) + "-" + df2["Month"].astype(str) + "-01", format="%Y-%m-%d")

# Drop the 'Year' and 'Month' columns if no longer needed
df2.drop(["Year", "Month"], axis=1, inplace=True)  #axis=0 represents dropping row,axis=1 represents dropping column,inplace replaces the original fi

# Display the resulting DataFrame
print(df2)
```
